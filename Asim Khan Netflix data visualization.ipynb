```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix_df=pd.read_csv("C:\\Users\\Asim\\Desktop\\Kaggle\\netflix_titles.csv")
```

```python
netflix_df.head()
```

```python
plt.figure(figsize=(15,9))
plt.title('Percentation comparision between TV Shows and Movies on Netflix')
g = plt.pie(netflix_df.type.value_counts(),explode=(0.050,0.025), labels=netflix_df.type.value_counts().index, colors=['yellow','green'],autopct= '%1.1f%%', startangle=180)
plt.show()
```

```python
filtered_directors = netflix_df[netflix_df.director != 'No Director'].set_index('title').director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
plt.figure(figsize=(14,6))
plt.title('Top 20 Director with most content directed')
sns.countplot(y = filtered_directors, order=filtered_directors.value_counts().index[:20], palette='Reds')
plt.show()
```

```python
filtered_genres = netflix_df.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);
plt.figure(figsize=(10,10))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:50], color = 'Green' )
plt.title('Top 50 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()
```
